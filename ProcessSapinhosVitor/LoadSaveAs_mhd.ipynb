```python
import matplotlib.pyplot as plt
import numpy as np
import SimpleITK as sitk
from PIL import Image
```

```python
'''
Função para leitura de uma imagem (tipo bmp) individual
fileName -  nome do arquivo
'''
def readBMPImgFromNRECON(fileName):
    img = Image.open(fileName)
    img = np.array(img)
    return img
```

```python
'''
Função para leitura de um diretorio (ou subconjunto) de imagens de um diretorio contendo imagens BMP
folderName - nome (path) do diretorio
fileNamePrefix - prefixo do nome do arquivo de imagem  
initImg - indice da imagem inicial
finalImg - indice da imagem final
step -  paso em grau entre as imagens, geralmente 1 
'''
def readBMPImgFolder(folderName, fileNamePrefix, initImg, finalImg, step):
    fileName = rf'{folderName}\{fileNamePrefix}{initImg:04d}.bmp'
    imgs = readBMPImgFromNRECON(fileName)
    for i in range(initImg+step, finalImg, step):
        if (i - initImg) % 50 == 0:
            print(f'Processadas {i - initImg} de {finalImg-initImg} imagens.')
        fileName = rf'{folderName}\{fileNamePrefix}{i:04d}.bmp'
        img = readBMPImgFromNRECON(fileName)
        imgs = np.dstack((imgs, img))
    return imgs.transpose((2, 0, 1))
```

```python
'''
Função para imprimir os detalhes de uma imagem após a leitura 
img - matriz da imagem
desc -  comentario opcional sobre a imagem
'''
def printImgData(img, desc = None):
    if desc!=None:
        print(f'{desc}:: ', end='')
    print(f'Shape:{img.shape}, Min:{img.min():.4e}, Max:{img.max():.4e}, Mean:{img.mean():.4e}, Type:{img.dtype}')
```

```python
folder = rf'F:\TC\Victor_Sapos\Reconstrucao 9092\slices'
```

```python
prefix='9092 9um 40kV_rec'
slices = readBMPImgFolder(folder, prefix, 385, 2380, 1)
printImgData(slices, 'Imagem BMP lida do NRECON')
```

```python
def salvar_como_mhd_raw(array_np, nome_base="imagem", espaco_voxel=(1.0, 1.0, 1.0)):

    # Converte para imagem SimpleITK
    imagem_sitk = sitk.GetImageFromArray(array_np.astype(np.uint16))  # ou np.uint16, conforme o tipo

    # Define o espaçamento dos voxels (em ordem Z, Y, X)
    imagem_sitk.SetSpacing(espaco_voxel)

    # Salva no formato .mhd/.raw
    sitk.WriteImage(imagem_sitk, f"{nome_base}.mhd")
```

```python
salvar_como_mhd_raw(slices, nome_base=rf"{folder}\MHD_RAW\TC_Sapinho_9092", espaco_voxel=(1.0, 1.0, 1.0))
```
